```python
#MACD (média da divergência de convergência média móve)
#Stochastics (Oscilador Estocástico para definir os movimentos dos preços das ações.)
#Average True Range(serve para mensurar a volatilidade do mercado, porém sem indicar a direção.)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
pd.options.mode.chained_assignment = None
import os
```

```python
def MACD(df,period1,period2,periodSignal):
    #Formula MACD=EMA de 12 períodos - EMA de 26 períodos
    EMA1 = pd.DataFrame.ewm(df,span=period1).mean()
    EMA2 = pd.DataFrame.ewm(df,span=period2).mean()
    MACD = EMA1-EMA2
    
    Signal = pd.DataFrame.ewm(MACD,periodSignal).mean()
    
    Histogram = MACD-Signal
    
    return Histogram

def stochastics_oscillator(df,period):
    #Formula-%K = (preço de fechamento – menor mínima)/(maior alta – menor mínima)*100
    #%D = Média Móvel Simples de 3 dias da %K
    
    l, h = pd.DataFrame.rolling(df, period).min(), pd.DataFrame.rolling(df, period).max()
    k = 100 * (df - l) / (h - l)
    return k


def ATR(df,period):
    '''
    Metodo A: diferença entre as maximas e minimas
    '''
    df['H-L'] = abs(df['MAXIMA']-df['MINIMA'])
    df['H-PC'] = abs(df['MAXIMA']-df['FECHAMENTO'].shift(1))
    df['L-PC'] = abs(df['MINIMA']-df['FECHAMENTO'].shift(1))
    TR = df[['H-L','H-PC','L-PC']].max(axis=1)
    return TR.to_frame()
```

```python
dataset = pd.read_excel(os.path.join('EUR/USD',r"C:\Users\edno2\Desktop\Rede_Binária\outros\DAT_XLSX_EURUSD_M1_201910.xlsx"))
dataset.head(-1)
```

```python
def dividindo_banco(tick):
    n=0
    max=[]
    min=[]
    for c in range(0,len(dataset)-(len(dataset)%tick),tick):
        max.append(dataset.iloc[c,1])
        min.append(dataset.iloc[c,1])
        for i in range(c,c+tick):     
            if dataset.iloc[i,1]>max[n]:
                max[n]=dataset.iloc[i,2]    
            if dataset.iloc[i,1]<min[n]:
                min[n]=dataset.iloc[i,3]          
        n=n+1
    return max,min

def abertu_fechamento(tick):
    data=[]
    aber=[]
    fec=[]
    for c in range(0,len(dataset)-(len(dataset)%tick),tick):
        data.append(dataset.iloc[c,0])
        aber.append(dataset.iloc[c,1])
        fec.append(dataset.iloc[c+tick-1,4])
        
    return data,aber,fec
        
def dividindo_time(tick):
    max,min= dividindo_banco(tick)
    data,aber,fec=abertu_fechamento(tick)  
    return data,aber,max,min,fec
   
    
novodata=dividindo_time(5)    
pc=pd.DataFrame(novodata)
pc=pc.T
pc.rename(columns={0:'DATA'}, inplace=True)
pc.rename(columns={1:'ABERTURA'}, inplace=True)
pc.rename(columns={2:'MAXIMA'}, inplace=True)
pc.rename(columns={3:'MINIMA'}, inplace=True)
pc.rename(columns={4:'FECHAMENTO'}, inplace=True)
pc.head(-1)
```

```python
def binario():# 3 subir 2 doji 1 descer
    bi=[0]
    for c in range(1,len(pc)):
        if pc.iloc[c,4]>pc.iloc[c-1,4]:
            bi.append(3)
        elif pc.iloc[c,4]<pc.iloc[c-1,4]:
            bi.append(1)
        elif pc.iloc[c,4]==pc.iloc[c-1,4]:
            bi.append(2)
    return bi

bi=pd.DataFrame(binario())
bi.rename(columns={0:'BINARIO'}, inplace=True)
bi.head(-1)



           
```
